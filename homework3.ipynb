```python
from math import sqrt
import matplotlib.pyplot as plt
import math
x=[]
x1=[]
y1=[]
y2=[]
y3=[]
y4=[]
r = 1

for i in range(10000):
  x_ = -i/10000
  x.append(x_)
  y1.append(math.sqrt(r**2 - x_**2))
  y2.append(-math.sqrt(r**2 - x_**2))
for i in range(10000):
  x_ = i/10000
  x1.append(x_)
  y3.append(math.sqrt(r**2 - x_**2))
  y4.append(-math.sqrt(r**2 - x_**2))
plt.plot(x,y1)
plt.plot(x,y2) 
plt.plot(x1,y3)
plt.plot(x1,y4) 
plt.xlabel("x")
plt.ylabel("y")
```

```python
from math import sqrt
import matplotlib.pyplot as plt
import math
x=[]
y1=[]
y2=[]
r = 0.5
for i in range(100):
  x_ = i/100
  x.append(x_)
  y1.append(math.sqrt(r**2 + x_**2))
  y2.append(-math.sqrt(r**2 + x_**2))
plt.plot(x,y1)
plt.plot(x,y2) 
plt.xlabel("x")
plt.ylabel("y")
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
Z2 = X + 2*Y + 50
Z1 = X + 2*Y
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
